```python
### check unique() ###

for i in df_oct:
  print(f'{i}\n',df_oct[i].unique(),'\n')
```

```python
### check value_counts() ###

for i in df_oct:
  print(f'{i}\n',df_oct[i].value_counts(dropna = False),'\n')
```

```python
#### Sum "bv_transaction", count "id", sum "nb_articles" by "family" #####
#### New columns that have % and Avg ###

df_oct_groupby = df_oct.groupby('family', as_index = False).agg({'bv_transaction':'sum',
                                                                 'id':'count',
                                                                 'nb_articles':'sum'})

#### create extra columns '% Transaction" #####
df_oct_groupby['% Transaction'] = df_oct_groupby['bv_transaction']/sum(df_oct_groupby['bv_transaction'])*100
df_oct_groupby['Avg Transaction'] = df_oct_groupby['bv_transaction']/(df_oct_groupby['id'])

#### create extra columns '% ID" #####
df_oct_groupby['% ID'] = df_oct_groupby['id']/sum(df_oct_groupby['id'])*100

#### create extra columns '% Article" #####
df_oct_groupby['% Article'] = df_oct_groupby['nb_articles']/sum(df_oct_groupby['nb_articles'])*100

#### create extra columns 'Mean Article" #####
df_oct_groupby['Avg Article by ID'] = df_oct_groupby['nb_articles']/(df_oct_groupby['id'])


df_oct_groupby.sort_values(by='bv_transaction', ascending = False)
```

```python
#### Fees, id by Score ####

df_oct_groupby = df_oct.groupby('score', as_index = False).agg({'shipping_fees':'sum',
                                                                'bv_transaction':'sum',
                                                                'id':'count',
                                                                'nb_articles':'sum'})

#### create extra columns 'Avg Fee' #####
# df_oct_groupby['% Transaction'] = df_oct_groupby['bv_transaction']/sum(df_oct_groupby['bv_transaction'])*100
df_oct_groupby['Avg Fee by ID'] = df_oct_groupby['shipping_fees']/(df_oct_groupby['id'])

#### create extra columns '% Fees" #####
df_oct_groupby['% Avg Fee'] = df_oct_groupby['shipping_fees']/sum(df_oct_groupby['shipping_fees'])*100

#### create extra columns 'Avg Article" #####
df_oct_groupby['Avg Fee by Transaction'] = df_oct_groupby['shipping_fees']/(df_oct_groupby['bv_transaction'])*100 ### this one I think doesn't make much sense to show ####

# #### create extra columns 'Mean Article" #####
# df_oct_groupby['Avg Article by ID'] = df_oct_groupby['nb_articles']/(df_oct_groupby['id'])


df_oct_groupby.sort_values(by='score', ascending = False)
```

```python
### check if the delivery is influenced on score ####

df_oct_groupby = df_oct.groupby('is_mf', as_index = False).agg({'score':'mean', 'nb_articles':'mean'})
df_oct_groupby.sort_values(by='score', ascending = False)
```

```python
### check if the delivery is influenced on score ####

df_oct_groupby = df_oct.groupby('csat_presales', as_index = False).agg({'score':'mean',
                                                                'nb_articles':'mean'})
df_oct_groupby.sort_values(by='csat_presales', ascending = False)
```
